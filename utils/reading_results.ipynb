## Sweep results

```python
import pandas as pd
import ast

# Read the file and parse each row as a dictionary
#file_path = '../data/results/task-il/seq-mnist/supcon/logs.txt'
#file_path = '../data/results/task-il/seq-cifar10/supcon/logs.txt'
file_path = '../data/results/task-il/seq-tinyimg/supcon/logs.txt'
NUM_TASKS = 10

with open(file_path, 'r') as file:
    data = [ast.literal_eval(line.strip()) for line in file]

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data)
df.head()
```

```python
df.columns
```

```python
df_sorted = df.sort_values(by=[f'accmean_task{NUM_TASKS}'], ascending=[False])
print(df_sorted)
```

```python
for t in range(1, NUM_TASKS+1):
    accuracy_columns = [f'accuracy_{i}_task{t}' for i in range(1, t+1)]
    df_sorted[f'avg_acc_task{t}'] = df_sorted[accuracy_columns].mean(axis=1)

print(df_sorted)
```

```python
for i in range(1, NUM_TASKS+1):
    df_sorted[f'bt_{i}'] = df_sorted[f'accuracy_{i}_task{NUM_TASKS}'] - df_sorted[f'accuracy_{i}_task{i}']

print(df_sorted)
```

```python
list(df_sorted.columns)
```

```python
columns_to_print = ['lr', 'temperature', 'alpha','accmean_task10'] \
    + [f'bt_{i}' for i in range(1, NUM_TASKS+1)] \
        + [f'accmean_task{i}' for i in range(1, NUM_TASKS+1)] 
```

```python
top_ten_asym_true = df_sorted[df_sorted['asym'] == True].head(10)
top_ten_asym_true = top_ten_asym_true[columns_to_print]
print(top_ten_asym_true)
```

```python
top_ten_asym_true = df_sorted[df_sorted['asym'] == False].head(10)
top_ten_asym_true = top_ten_asym_true[columns_to_print]
print(top_ten_asym_true)
```
