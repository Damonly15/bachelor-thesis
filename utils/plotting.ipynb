```python
import pandas as pd
import ast

# Read the file and parse each row as a dictionary
file_path = '../data/results/task-il/seq-mnist/supcon/logs.txt'
with open(file_path, 'r') as file:
    data = [ast.literal_eval(line.strip()) for line in file]

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data)
print(df)
```

```python
df.columns
```

```python
df_sorted = df.sort_values(by=['accmean_task5'], ascending=[False])
print(df_sorted)
```

```python
for t in range(1, 6):
    accuracy_columns = [f'accuracy_{i}_task{t}' for i in range(1, t+1)]
    df_sorted[f'avg_acc_task{t}'] = df_sorted[accuracy_columns].mean(axis=1)

print(df_sorted)
```

```python
for i in range(1, 6):
    df_sorted[f'bt_{i}'] = df_sorted[f'accuracy_{i}_task5'] - df_sorted[f'accuracy_{i}_task{i}']

print(df_sorted)
```

```python
df_sorted.columns
```

```python
top_ten_asym_true = df_sorted[df_sorted['asym'] == True].head(10)
columns_to_print = [
    'lr', 'temperature', 'alpha','accmean_task5', 'bt_1', 'bt_2',
       'bt_3', 'bt_4', 'bt_5',
    'accmean_task1', 'accmean_task2', 'accmean_task3', 'accmean_task4', 
    'accuracy_1_task1', 'accuracy_1_task2', 'accuracy_2_task2', 'accuracy_1_task3',
    'accuracy_2_task3', 'accuracy_3_task3', 'accuracy_1_task4', 'accuracy_2_task4',
    'accuracy_3_task4', 'accuracy_4_task4', 'accuracy_1_task5', 'accuracy_2_task5',
    'accuracy_3_task5', 'accuracy_4_task5', 'accuracy_5_task5', 'forward_transfer',
]
top_ten_asym_true = top_ten_asym_true[columns_to_print]
print(top_ten_asym_true)
```
